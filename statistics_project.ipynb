```python
for col in ['fare_amount', 'trip_distance', 'duration']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1

    lower_bound = q1-1.5*IQR
    upper_bound = q3+1.5*IQR

    df = df[(df[col]>=lower_bound)&(df[col]<=upper_bound)]

df
```

```python
# Passenger Count Analysis
passenger_count = df.groupby(['payment_type', 'passenger_count'])[['passenger_count']].count()
passenger_count.rename(columns={'passenger_count':'count'}, inplace= True)
passenger_count.reset_index(inplace=True)
```

```python
# Horizontal Stacked Bar Plot
fig, ax = plt.subplots(figsize=(20, 6))
passenger_plot_df.plot(
    x='payment_type',
    kind='barh',
    stacked=True,
    ax=ax,
    color=['#FA643F', '#FFBCAB', '#CBB2B2', '#F1F1F1', '#FD9F9F']
)

# Add percentage text
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    ax.text(
        x + width / 2,
        y + height / 2,
        f'{width:.0f}%',
        ha='center',            # horizontalalignment in short ha
        va='center'             # verticalalignment in short va
    )

plt.title('Passenger Count Distribution by Payment Type')
plt.xlabel('Percentage')
plt.ylabel('Payment Type')
plt.show()
```

# Hypothesis Testing
**Null hypothesis (H0):** There is no difference in average fare between customers who use credit cards and customers who use cash

**Alternative hypothesis (H1):** There is a difference in average fare between customers who use credit cards and customers who use cash

```python
# QQ Plot for 'fare_amount' using the original DataFrame
sm.qqplot(df['fare_amount'], line='45')
plt.title('QQ Plot of Fare Amount')
plt.show()
```

```python
# Fare amounts for each group
card_fares = df[df['payment_type'] == 'Card']['fare_amount']
cash_fares = df[df['payment_type'] == 'Cash']['fare_amount']
```

```python
# Perform t-test
t_stat, p_value = st.ttest_ind(card_fares, cash_fares, equal_var=False)  # set equal_var=True if variances are equal

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

# Regression Analysis

```python
# Independent variable (X) - duration
X = df[['duration']]

# Dependent variable (Y) - fare amount
Y = df['fare_amount']
```

```python
# Initializing and fitting the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Making predictions
Y_pred = model.predict(X)
```

```python
# Intercept and Coefficient
intercept = model.intercept_
coefficient = model.coef_[0]

# R-squared value to assess the fit
r_squared = model.score(X, Y)
```

```python
# Plotting the results
plt.scatter(X, Y, color='blue', label='Actual Fare')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Predicted Fare')
plt.xlabel('Duration (minutes)')
plt.ylabel('Fare Amount')
plt.legend()
plt.title('Linear Regression: Fare Amount vs Duration')
plt.show()

print(f"Intercept: {intercept}, Coefficient: {coefficient}, r_squared: {r_squared}")
```
